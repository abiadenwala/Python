```
# Get from Widgets
JobID = dbutils.widgets.get("JobID")
RunID = dbutils.widgets.get("RunID")
```

```
localIp = os.environ.get("SPARK_LOCAL_IP")
uiPort = spark.conf.get("spark.ui.port")
url="http://"+localIp+":"+uiPort+"/api/v1/applications/"
# Get application details
applications = requests.get(url).json()

#loop through list and find value for key "id"
print(applications)
# initialize an empty list
applicationIDs = []
for app in applications:
    applicationIDs.append(app['id'])
ApplicationID = app['id']
```

```
jobsurl = url + ApplicationID + "/jobs"
JobDetails = requests.get(jobsurl).json()
Job_UniqID = 'job-{}-run-{}'.format(JobID, RunID)
search_str = '_job-{}-run-{}-'.format(JobID, RunID)
#print (requests.get(jobsurl).text)
validJobValue = None
stageIds = None
for item in JobDetails:
  jobGrp = item['jobGroup']
  if search_str in jobGrp:
    validJobValue = item['jobGroup']
    stageIds = item['stageIds']
    print(stageIds)
    break

if stageIds == None:
  print (Job_UniqID + " not found or has no stages")
```
